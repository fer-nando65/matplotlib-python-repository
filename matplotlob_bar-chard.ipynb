```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company=['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue= [90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company, revenue)
plt.bar(company, profit)
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.title('Tech Stock')
plt.ylabel('Revenue')

plt.xticks(xpos, company) #replacing xpos label into company label
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='Profit')
plt.legend()
plt.grid()
```

```python
plt.title('Tech Stock')

plt.yticks(xpos, company) #replacing xpos label into company label
plt.barh(xpos-0.2, revenue, height = 0.4, label='Revenue') #horizontal bar
plt.barh(xpos+0.2, profit, height= 0.4, label='Profit')
plt.legend()
plt.grid()
```
