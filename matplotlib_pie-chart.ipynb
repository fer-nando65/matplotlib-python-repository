```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600, 300,410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill1","Car ", "other utilities"]
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5, explode=[0,.2,0,0,0], shadow=True, autopct='%0.2f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5, explode=[0,.2,0,0,0], shadow=True, autopct='%0.2f%%', startangle=90)
plt.show()
```

## Saving pie chart into png file

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5, explode=[0,.2,0,0,0], shadow=True, autopct='%0.2f%%', startangle=90)

plt.savefig('piechart', bbox_inches='tight', pad_inches=2)
```
